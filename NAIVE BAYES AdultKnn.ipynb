```python
import numpy as np
import pandas as pd
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
a=pd.read_csv("AdultKNN.csv")
```

```python
a.head()
```

```python
a=a.replace({"?":np.nan})
```

```python
a.info()
```

```python
a.isnull().sum()
```

```python
a.workclass=a.workclass.fillna("Private")
```

```python
a.occupation=a.occupation.fillna("Prof-specialty")
```

```python
a["native-country"]=a["native-country"].fillna("United-States")
```

```python
a.income=a.income.replace({"<=50K":1,">50K":0})
```

```python
a.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a[a.select_dtypes(include=["object"]).columns]=a[a.select_dtypes(include=["object"]).columns].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test=train_test_split(a,test_size=0.2)
```

```python
a_train_x=a_train.iloc[:,:-1]
a_train_y=a_train.iloc[:,-1]
```

```python
a_test_x=a_test.iloc[:,:-1]
a_test_y=a_test.iloc[:,-1]
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
n=MultinomialNB()
```

```python
n.fit(a_train_x,a_train_y)
```

```python
pred=n.predict(a_test_x)
pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(a_test_y,pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(a_test_y,pred)
```
