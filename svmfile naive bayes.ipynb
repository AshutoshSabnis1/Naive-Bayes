```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a[a.select_dtypes(include=["object"]).columns]=a[a.select_dtypes(include=["object"]).columns].apply(le.fit_transform)
```

```python
# Negative values run the fit transform function individually
a["cons.conf.idx"]=le.fit_transform(a["cons.conf.idx"])
a["emp.var.rate"]=le.fit_transform(a["emp.var.rate"])
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test=train_test_split(a,test_size=0.2)
```

```python
a_train_x=a_train.iloc[:,:-1]
a_train_y=a_train.iloc[:,-1]
```

```python
a_test_x=a_test.iloc[:,:-1]
a_test_y=a_test.iloc[:,-1]
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
n=MultinomialNB()
```

```python
n.fit(a_train_x,a_train_y)
```

```python
pred=n.predict(a_test_x)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(a_test_y,pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(a_test_y,pred)
```

```python
from sklearn.svm import SVC
s=SVC(kernel="poly",degree=5)
s.fit(a_train_x,a_train_y)
```

```python
pred=s.predict(a_test_x)
```

```python
pred
```

```python
confusion_matrix(a_test_y,pred)
```

```python
accuracy_score(a_test_y,pred)
```
